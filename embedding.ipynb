```python

from tensorflow.keras.preprocessing.text import one_hot # type: ignore|
```

```python
# sentence

sent= [
    'the glass of milk',
    'the glass of juice',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
# Define the vocabulary size
vocab_size =10000
```

```python
## One hot Representation
one_hot_repr = [one_hot(words ,vocab_size) for words in sent]
one_hot_repr
```

```python
# word embedding representation

from tensorflow.keras.layers import Embedding # type: ignore
from tensorflow.keras.preprocessing.sequence import pad_sequences # type: ignore
from tensorflow.keras.models import Sequential # type: ignore
```

```python
import numpy as np
```

```python
sent_length =8
embedded_docs = pad_sequences(one_hot_repr, padding='pre', maxlen=sent_length)
print(embedded_docs)
```

```python
# feature representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_length))
model.compile(optimizer='adam',loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
